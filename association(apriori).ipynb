```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Market_Basket_Optimisation.csv", header=None)
```

```python
data
```

```python
transcation = []
```

```python
for i in range(0,7501):
  transcation.append([str (data.values[i, j]) for j in range(0,20)])
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions = transcation, min_support =0.003 , min_confidence = 0.2, min_lift = 3 , min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
results
```

```python
print(rules)
```

```python
resultsdata = pd.DataFrame(results)
```

```python
print(resultsdata.columns)
```
